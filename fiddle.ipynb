```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
feature = pd.read_csv('data/feature.csv', index_col=0)
price = pd.read_csv('data/price.csv', index_col=0)
```

```python
df = pd.concat([feature, price], axis=1)
```

```python
df.corr()
```

```python
hdb = df[df['type_hdb']==1]
condo = df[df['type_condo'] ==1]
```

```python
condo.corr()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
```

```python
Y = condo['price_sqft']
Y
```

```python
for i in [2,3,4]:
    data = df
    poly = PolynomialFeatures(degree=i)
    X = poly.fit_transform(data.iloc[:,:-1])
    Y = data['price_sqft']
    kfold = KFold(n_splits=5)
    
    diff_max = []
    diff_min = []
    pred_max = []
    pred_min = []
    err = []

    for train_idx, test_idx in kfold.split(Y):
        train_x = X[train_idx]
        train_y = Y[train_idx]
        test_x = X[test_idx]
        test_y = Y[test_idx]

        linreg = LinearRegression()
        linreg.fit(train_x, train_y)
        y_pred = linreg.predict(test_x)
        err.append(np.sqrt(mean_squared_error(y_pred, test_y)))
        pred_max.append(np.max(y_pred))
        pred_min.append(np.min(y_pred))
        diff_max.append(np.max(y_pred - test_y))
        diff_min.append(np.min(y_pred - test_y))
        print('Done')

    
    print()
    print("Pred max: ", np.max(pred_max))
    print("Pred min: ", np.min(pred_min))
    print("Diff max: ", np.max(diff_max))
    print("Diff min: ", np.min(diff_min))
    print()
    print(err)
    print(np.mean(err))
    print()
```

```python
from sklearn.linear_model import Ridge

for i in [2,3,4]:
    data = df
    poly = PolynomialFeatures(degree=i)
    X = poly.fit_transform(data.iloc[:,:-1])
    Y = data['price_sqft']
    kfold = KFold(n_splits=5)
    
    diff_max = []
    diff_min = []
    pred_max = []
    pred_min = []
    err = []

    for train_idx, test_idx in kfold.split(Y):
        train_x = X[train_idx]
        train_y = Y[train_idx]
        test_x = X[test_idx]
        test_y = Y[test_idx]

        linreg = Ridge(alpha=10)
        linreg.fit(train_x, train_y)
        y_pred = linreg.predict(test_x)
        err.append(np.sqrt(mean_squared_error(y_pred, test_y)))
        pred_max.append(np.max(y_pred))
        pred_min.append(np.min(y_pred))
        diff_max.append(np.max(y_pred - test_y))
        diff_min.append(np.min(y_pred - test_y))
        print('Done')

    
    print()
    print("Pred max: ", np.max(pred_max))
    print("Pred min: ", np.min(pred_min))
    print("Diff max: ", np.max(diff_max))
    print("Diff min: ", np.min(diff_min))
    print()
    print(err)
    print(np.mean(err))
    print()
```
